```python
# In-Class Assignment 1: Income Statement using Simple Income Statement Calculation:
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))
cogs = 0.4 * sales
gross_profit = sales - cogs
tax = gross_profit * (tax_rate / 100)
net_income = gross_profit - tax
print(f"Sales: ${sales:,.1f}")
print(f"Cost of Goods Sold: ${cogs:,.2f}")
print(f"Gross Profit: ${gross_profit:,.2f}")
print(f"Taxes: ${tax:,.2f}")
print(f"Net Income: ${net_income:,.2f}")
```

```python
# In-Class Assignment 6: Rock-Paper-Scissors:
import random

play = input("Do you want to play a game? (y/n): ")
while (play == 'y'):
  user_input = input("Please choose rock, paper or scissors: ")
  user_input = user_input.lower()

  while user_input != "rock" and user_input != "paper" and user_input != "scissors":
    user_input = input("Please choose rock, paper or scissors: ")
    user_input = user_input.lower()

  computer_choice = random.choice(["rock", "paper", "scissors"])
  print(f"Computer chose {computer_choice}")
  print(f"You chose {user_input}")

  if (user_input == "rock" and computer_choice == "scissors") or (user_input == "paper" and computer_choice == "rock") or (user_input == "scissors" and computer_choice == "paper"):
    print("You win!")
  elif user_input == computer_choice:
    print("It's a tie!")
  else:
    print("You lose!")

  play = input("Do you want to play again? (y/n): ")

```

```python
# In-Class Assignment 11: Parking Garage: calcFee() function:
def calcFee(hours):
  base_fee = 5.0
  hourly_rate = 2.5
  fee = base_fee + (hours) * hourly_rate
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0
  return fee
def main():
  hours = int(input("Enter hours: "))
  print(calcFee(hours))
main()
```

```python
# In-Class Assignment 12: Recursive Function: String Reversal:
def reverse(string):
  if len(string) == 0:
    return string
  else:
    return reverse(string[1:]) + string[0]

def main():
  string = input("Enter a string: ")
  print(reverse(string))

main()
```

```python
# In-Class Assignment 13: ROI Calculator:
def roi(cost, revenue):
  if not isinstance(cost, (int, float)) or not isinstance(revenue, (int, float)):
    return "Error! Both cost and revenue must be numeric types (int or float)."
  if cost <= 0:
        return "Cost must be greater than 0."
  return (revenue - cost) / cost

cost = float(input("Please enter cost: "))
revenue = float(input("Please enter revenue: "))
print(roi(cost, revenue))
```

```python
# In-Class Assignment 14: Loan Payment Calculator:
def loan_payment(principal, annual_rate, years):
  if principal <= 0 or annual_rate <= 0 or years <= 0:
    return "Error! Please enter a positive number."

  r = annual_rate / 12
  n = years * 12
  payment = ((r * principal) / (1 +r) ** -n)

  return payment

principal = float(input("Please enter the principal: "))
annual_rate = float(input("Please enter the annual rate: "))
years = float(input("Please enter the years: "))

print(loan_payment(principal, annual_rate, years))
```

```python
# In-Class Assignment 15: Quality Control Manager (exceptions):
def getValues(n):
  values = []
  for i in range(n):
    while True:
      try:
        value = float(input(f"Please enter value {i + 1}: "))
        if value < 0:
          print("Error! Please enter a positive number.")
        else:
          values.append(value)
          break
      except (ValueError, TypeError):
        print("Error! Please enter a numeric value.")
  return values

def calcAverage(alist):
  trimmed = alist[3:]
  return sum(trimmed) / len(trimmed)

while True:
  try:
    n = int(input("How many values? "))
    if n <= 3:
      print("Error! Please enter more than 3 values.")
    else:
      break
  except ValueError:
    print("Error! Please enter a whole number.")

values = getValues(n)
print(f"Average (excluding first 3): {calcAverage(values)}")


```

```python
# In-Class Assignment 16: Greeting Class:
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(f"{self.first}")

  def __str__(self):
    return f"{self.first} {self.last}"

g = Greeting("John", "Smith")
g.display(True)
g.display(False)
print(g)
```

```python
# In-Class Assignment 17: Student class:
class Student:
  def __init__(self, name, number, courses):
    if not isinstance(name, str):
        print("Error! Name must be a string.")
    self.name = name
    self.number = number
    self.courses = courses

  def get_name(self):
    return self.name

  def get_courses(self):
    return self.courses

  def set_name(self, new_name):
    if isinstance(new_name, str):
      self.name = new_name
    else:
      print("Error! Name must be a string.")

  def __len__(self):
    return len(self.courses)

  def enroll(self, course):
    if course not in self.courses:
      self.courses.append(course)
    else:
      print(f"{self.name} is already enrolled in {course}.")
```
