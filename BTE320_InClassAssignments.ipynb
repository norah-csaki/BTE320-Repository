```python
# In-Class Assignment 1: Income Statement using Simple Income Statement Calculation:
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax rate (%): "))
cogs = 0.4 * sales
gross_profit = sales - cogs
tax = gross_profit * (tax_rate / 100)
net_income = gross_profit - tax
print(f"Sales: ${sales:,.1f}")
print(f"Cost of Goods Sold: ${cogs:,.2f}")
print(f"Gross Profit: ${gross_profit:,.2f}")
print(f"Taxes: ${tax:,.2f}")
print(f"Net Income: ${net_income:,.2f}")
```

```python
# In-Class Assignment 2: Inventory Management:
import math

demand = float(input("Please enter the demand: "))
order_cost = float(input("Please enter the order cost: "))
holding_cost = float(input("Please enter the holding cost: "))

EOQ = math.sqrt((2 * demand * order_cost) / holding_cost)
print(f"The EOQ is {EOQ:.2f}")
```

```python
# In-Class Assignment 4: Parking garage (for-loop edition):
base_fee = 5.0
hourly_rate = 2.5

for i in range(1, 9):
  fee = base_fee + (i) * hourly_rate
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0
  print(f"{i} {fee}")
```

```python
# In-Class Assignment 5: Movie Theater (for-loop edition):
init_attendees = 20
fixed_cost = 200

print("Ad$ Profit")
for i in range(0, 201, 25):
  revenue = (init_attendees + 2 * round(i ** 0.5)) * 10
  costs = fixed_cost + i
  profit = revenue - costs
  print(float(i), float(profit))

```

```python
# In-Class Assignment 4: Parking Garage (while-loop edition):
base_fee = 5.0
hourly_rate = 2.5
i = 1

while i >= 1 and i <= 8:
  fee = base_fee + (i) * hourly_rate
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0
  print(f"{i} {fee}")
  i+=1
```

```python
# In-Class Assignment 6: Rock-Paper-Scissors:
import random

play = input("Do you want to play a game? (y/n): ")
while (play == 'y'):
  user_input = input("Please choose rock, paper or scissors: ")
  user_input = user_input.lower()

  while user_input != "rock" and user_input != "paper" and user_input != "scissors":
    user_input = input("Please choose rock, paper or scissors: ")
    user_input = user_input.lower()

  computer_choice = random.choice(["rock", "paper", "scissors"])
  print(f"Computer chose {computer_choice}")
  print(f"You chose {user_input}")

  if (user_input == "rock" and computer_choice == "scissors") or (user_input == "paper" and computer_choice == "rock") or (user_input == "scissors" and computer_choice == "paper"):
    print("You win!")
  elif user_input == computer_choice:
    print("It's a tie!")
  else:
    print("You lose!")

  play = input("Do you want to play again? (y/n): ")

```

```python
# In-Class Assignment 7: Append items to list iteratively (while loop):
numbers = []

while True:
  num = float(input("Please enter a numerical input: "))
  if num == 0:
    break
  numbers.append(num)

numbers.sort(reverse=True)
print(numbers)
```
